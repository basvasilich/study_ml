```python
import math
import numpy as np
from scipy.optimize import minimize

def fun(x):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5.0 * math.exp(-x / 2.0)

res1 = minimize(fun, [2], method="BFGS")
res2 = minimize(fun, [30], method="BFGS")

res = [round(res1.fun, 2), round(res2.fun, 2)]

f1 = open("sub1.txt","w")
f1.write(f"{res[0]} {res[1]}")
f1.close()                                  

res
```

```python
%matplotlib inline
from matplotlib import pylab as plt

x = np.arange(1, 30, 0.1)
y1 = [fun(_) for _ in x]

plt.plot(x, y1)
plt.show()
```

```python
from scipy.optimize import differential_evolution

res3 = differential_evolution(fun, [(1, 30)])

f2 = open("sub2.txt","w")
f2.write(f"{round(res3.fun, 2)}")
f2.close()  
```

```python
def g(x):
    return int(fun(x))

y2 = [g(_) for _ in x]
plt.plot(x, y2)
plt.show()
```

```python
res4 = minimize(g, [30], method="BFGS")
res5 = differential_evolution(g, [(1, 30)])

f2 = open("sub3.txt","w")
f2.write(f"{res4.fun} {res5.fun}")
f2.close()  

[res4.fun, res5.fun]
```
