```python
import math
import numpy as np
from scipy.optimize import minimize

def f(x):
    return math.sin(x / 5.0) * math.exp(x / 10.0) + 5.0 * math.exp(-x / 2.0)
res1 = minimize(f, [2], method="BFGS")
res2 = minimize(f, [30], method="BFGS")

res = [round(f(res1.x),2), round(f(res2.x), 2)]
f = open("sub1.txt","w")
f.write(f"{res[0]} {res[1]}")
f.close()                                  
res
```

```python
%matplotlib inline
from matplotlib import pylab as plt

x = np.arange(1, 30, 0.1)
y = [f(_) for _ in x]

plt.plot(x, y)
plt.show()
```
